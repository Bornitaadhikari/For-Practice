Set Colors to Label

```python
x=[80,85,90,95,100,105]
y=[240,250,260,280,290,300]
plt.plot(x,y)
plt.title("Our Plot",color='r',size=20)
plt.xlabel("X axis",color='b',size=25)
plt.ylabel("Y axis",color='c',size=20);
```

POsition to title

```python
x=[80,85,90,95,100,105]
y=[240,250,260,280,290,300]
plt.plot(x,y)
font1 = {'family': 'serif','color':'blue','size':20}
font2 = {'family': 'serif','color':'red','size':10}
plt.title("Our Plot",color='r',fontdict=font1)
plt.xlabel("X axis",color='b',fontdict=font2)
plt.ylabel("Y axis",color='c',size=20);
```

POsition to title

```python
x=[80,85,90,95,100,105]
y=[240,250,260,280,290,300]
plt.plot(x,y)
font1 = {'family': 'serif','color':'blue','size':20}
font2 = {'family': 'serif','color':'red','size':10}
plt.title("Our Plot",color='r',fontdict=font1,loc='left')
plt.xlabel("X axis",color='b',fontdict=font2)
plt.ylabel("Y axis",color='c',size=20);
```

Add Grid Lines to a PLot

```python
x=[80,85,90,95,100,105]
y=[240,250,260,280,290,300]
plt.plot(x,y)
font1 = {'family': 'serif','color':'blue','size':20}
font2 = {'family': 'serif','color':'red','size':10}
plt.title("Our Plot",color='r',fontdict=font1,loc='left')
plt.xlabel("X axis",color='b',fontdict=font2)
plt.ylabel("Y axis",color='c',size=20);
plt.grid()

```

Specify Which Lines to Plot

```python
x=[80,85,90,95,100,105]
y=[240,250,260,280,290,300]
plt.plot(x,y)
font1 = {'family': 'serif','color':'blue','size':20}
font2 = {'family': 'serif','color':'red','size':10}
plt.title("Our Plot",color='r',fontdict=font1,loc='left')
plt.xlabel("X axis",color='b',fontdict=font2)
plt.ylabel("Y axis",color='c',size=20);
plt.grid(axis='y')
```

Set Properties to Grid

```python
x=[80,85,90,95,100,105]
y=[240,250,260,280,290,300]
plt.plot(x,y)
font1 = {'family': 'serif','color':'blue','size':20}
font2 = {'family': 'serif','color':'red','size':10}
plt.title("Our Plot",color='r',fontdict=font1,loc='left')
plt.xlabel("X axis",color='b',fontdict=font2)
plt.ylabel("Y axis",color='c',size=20);
plt.grid(color='green',linestyle='--',linewidth=2)
```

# Creating Pie

```python
y=[35,25,25,15]
plt.pie(y);
```

```python
y=[35,25,25,15]
myl=['Apple','Banana','Orange','Mango']
plt.pie(y,labels=myl,startangle=90);
```

Explode

```python
y=[35,25,25,15]
myl=['Apple','Banana','Orange','Mango']
mye=[0.2,0,0,0]
plt.pie(y,labels=myl,startangle=90,explode=mye);
```

Shadow

```python
y=[35,25,25,15]
myl=['Apple','Banana','Orange','Mango']
mye=[0.2,0,0,0]
plt.pie(y,labels=myl,startangle=90,explode=mye,shadow=True);
```
